```python
import bs4
import pandas as pd
import requests

listings_url = 'https://www.truecar.com/used-cars-for-sale/listings/audi/location-lake-orion-mi/?searchRadius=5000'
listings_txt = requests.get(listings_url).text
soup = bs4.BeautifulSoup(listings_txt)

all_listings = soup.find(attrs={"data-test": "allVehicleListings"})

listings = (all_listings.find_all(class_= 'flex-column flex-sm-row card margin-top-3 card_1hx6mcr'))

# pulling initial fields for one record below
# print(listings[1].find('h4').get_text())
# print(listings[1].find(attrs={"data-test": "vehicleListingMileage"}).get_text())
# price
# print(listings[1].find('h4', class_= 'heading-2').get_text())
# print(listings[1].find(attrs={"data-test": "vehicle-listing-exterior-colors"}).get_text())
# print(listings[1].find(attrs={"data-test": "vehicle-listing-interior-colors"}).get_text())
# print(listings[1].find(attrs={"data-test": "vehicle-listing-location"}).get_text())
# get VIN from href
# print(listings[1].find('a')['href'])

# get each record; loop with list comprehensions within webpage
VIN = [listing.find('a')['href'] for listing in listings]
Model_Desc = [listing.find('h4').get_text() for listing in listings]
Price = [listing.find('h4', class_= 'heading-2').get_text() for listing in listings]
Mileage = [listing.find(attrs={"data-test": "vehicleListingMileage"}).get_text() for listing in listings]
Ext_Color = [listing.find(attrs={"data-test": "vehicle-listing-exterior-colors"}).get_text() for listing in listings]
Int_Color = [listing.find(attrs={"data-test": "vehicle-listing-interior-colors"}).get_text() for listing in listings]
Distance = [listing.find(attrs={"data-test": "vehicle-listing-location"}).get_text() for listing in listings]

# create the data frame using a dictionary
listings_tbl = pd.DataFrame(
    {'Vehicle_ID': VIN,
     'Model': Model_Desc,
     'Asking_Price': Price,
     'Miles_Driven': Mileage,
     'Exterior_Color': Ext_Color,
     'Interior_Color': Int_Color,
     'Listing_Location': Distance
    })

# print the data frame
print(listings_tbl)

# save data frame to csv file in working directory
listings_tbl.to_csv('Audi_Listings.csv')

# this is here for html review
# print(all_listings.prettify())
```
